```python
#Importing the libraries 
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import textwrap
plt.style.use('ggplot')


#Importing the dataset using CSV
dataset = pd.read_csv("C:\\Users\\Juhaina\\Documents\\\GitHub\\Olympic_Athletes\\athlete_events.csv",index_col=0)

#viewing the first 10 rows of the table
# note- the Index is not unique as one person can compete in multiple events- gotta replace the current index 
# I wont create a multi-index because a person can compete in multiple events in multiple years and cities blah blah
dataset.head(10)
```

```python
#Rest the index so that every row is unique 
dataset.reset_index(inplace=True)
dataset
```

```python
# Exploring the number of columns and number of rows
dataset.shape
```

```python
#Exploring the data types of the columns 
dataset.dtypes
```

```python
#Converting the format of age from float to int because no one is 19.7 years old... arguably 
dataset['Age'] = dataset['Age'].astype('Int64')
dataset
```

```python
#Lets also delete the Games column because 
dataset.drop('Games', axis=1, inplace=True)
```

```python
#Checking if there are duplicate rows, damn 1385 duplicates
dataset.loc[dataset.duplicated()]
```

```python
# getting rid of the duplicates 
dataset.drop_duplicates(inplace=True)
dataset
```

```python
#do we have any na's in the dataset? ofcourse we do, im not ashamed of my age, height or weight. just kidding im sure its something else
dataset.isna().sum()
```

```python
# lets checkout some basic info on the dataset. imagine being an olympian at 10 ??! SLAYYY
dataset.describe()
```

```python
#Speaking of age lets do some basic analysis
#lets look at the number of total olympians by age
#Well.... I guess i cant use my age as an excuse anymore---
Age= dataset['Age'].value_counts().head(30).plot(kind='bar', figsize=(42,28))
Age.spines['bottom'].set_color('0')
Age.spines['left'].set_color('0')
Age.set_facecolor("white")
plt.xticks(rotation =360, fontsize=40)
plt.yticks( fontsize=40)
plt.xlabel('Age',fontsize=40)
plt.grid(axis = 'y',color = 'grey', linestyle = '--', linewidth = 0.3)
```

```python
#Lets look at if there are any correlations in the data
# Looks like theres a strong correlation between height and weight 
correlation= dataset.dropna().corr()
correlation
```

```python
# lets visualize that im lazy 
sns.heatmap(correlation, annot=True)
```

```python
# lets vizualize that using a scatterplot 
# so basically,the taller you are, the more you weight, thats right im not fat... im tall!
dataset.plot(kind='scatter', 
             x='Height',
             y='Weight')
plt.show()
```

```python
#Okay now lets look at sex, not surprised, the number of females is a 1/3 of the number of males 
#Buttt this is overall so lets look at it by year 
Sex=dataset['Sex'].value_counts().plot(kind='bar')
Sex.spines['bottom'].set_color('0')
Sex.spines['left'].set_color('0')
Sex.set_facecolor("white")
plt.xticks(rotation =360, fontsize=20)
plt.yticks( fontsize=20)
plt.xlabel('Sex',fontsize=20)
plt.grid(axis = 'y',color = 'grey', linestyle = '--', linewidth = 0.3)
```

```python
#First, let us just look at the years this dataset contains 
sorted(dataset['Year'].unique())
```

```python
#Okay now lets see if that got better over the years
#Barely any females until 1912 sheeshhh..its getting better, not there yet.. maybe the male population is generally higher?
#A quick google search says that there are more men in the world...something to explore further hmm
#but it looks like we have seasons?
ax= sns.countplot(x='Year',hue='Sex', data= dataset,palette='Blues')
sns.set(rc = {'figure.figsize':(24,10)})
ax.set_facecolor("white")
ax.spines['bottom'].set_color('0')
ax.spines['left'].set_color('0')
```

```python
#Lets get the exact count of males and females by year, no females in 1896 and only 16 in 1904??? SHAME
#However, its lovely to see the gap reducing
dataset.groupby(['Year','Sex'])['Sex'].count()
```

```python
#Now lets investigate the seasonality
# so they only has summer seasons until 1920 and theyve been alternating since 1994
# and obviously the summer season has more olympians participating 
dataset.pivot_table(index='Year', columns=['Season'],values='Sex',aggfunc='count' )
```

```python
# lets count the number of medals won by year and set
#BUT THATS NOT AN ACCURATE REPRESENTATION AS THE TOTAL NUMBER OF MALES AND FEMALES is not equal SO WE'LL HAVE TO GET A %
# maybe not the best way to see it
SummerMedals=dataset[dataset['Season'].str.contains('Summer', na=False)]
SummerMedals= SummerMedals.pivot_table(index='Year', columns=['Sex'],values='Medal',aggfunc='count' )
SummerMedals=SummerMedals.plot(kind='bar', figsize=(50,28))
SummerMedals.spines['bottom'].set_color('0')
SummerMedals.spines['left'].set_color('0')
SummerMedals.set_facecolor("white")
plt.xticks(rotation =360, fontsize=40)
plt.yticks( fontsize=40)
plt.xlabel('Age',fontsize=40)
SummerMedals.grid(axis = 'y',color = 'grey', linestyle = '--', linewidth = 0.3)
SummerMedals.legend(loc='upper right',fontsize=36)
```
