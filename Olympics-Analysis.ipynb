```python
#Importing the libraries 
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import textwrap
import matplotlib as plt


#Importing the dataset using CSV
dataset = pd.read_csv("C:\\Users\\Juhaina\\Documents\\\GitHub\\Olympics-Analysis\\Olympic_Athletes\\athlete_events.csv",index_col=0)

#viewing the first 10 rows of the table
# note- the Index is not unique as one person can compete in multiple events- I will have to replace the current index 
# I wont create a multi-index because a person can compete in multiple events in multiple years and cities etc
dataset.head(10)
```

```python
#Reset the index so that every row is unique 
dataset.reset_index(inplace=True)
dataset
```

```python
# Exploring the number of columns and number of rows
dataset.shape
```

```python
#Exploring the data types of the columns 
dataset.dtypes
```

```python
#Converting the format of age from float to int because no one is 19.7 years old... arguably 
dataset['Age'] = dataset['Age'].astype('Int64')
dataset
```

```python
#Lets also delete the Games column
dataset.drop('Games', axis=1, inplace=True)
```

```python
#Checking if there are duplicate rows, damn 1385 duplicates
dataset.loc[dataset.duplicated()]
```

```python
# getting rid of the duplicates 
dataset.drop_duplicates(inplace=True)
dataset
```

```python
#do we have any na's in the dataset? ofcourse we do,the na's in medal is ok because not everyone wins a medal but the others?
dataset.isna().sum()
```

```python
# lets checkout some basic info on the dataset. imagine being an olympian at 10 ??! 
# The youngest person to compete was 10 years old ,the oldest person was 97 and the average is 25 years old
# The average height is 175cm, the shortest to compete was 127cm, and the tallest was 226
# The average weight is 70kgs, The lowest was 25kgs, the heaviest was 214kgs
# This dataset covers 120 years of olympics from 1896 to 2016
dataset.describe()
```

```python
# Filtering by the age to find out about the youngest Olympian, which is Dimitrios Loundras
YoungestAge =dataset.loc[dataset['Age'].idxmin()]
YoungestAge
```

```python
#Now lets look at the oldest 
Oldest=dataset.loc[dataset['Age'].idxmax()]
Oldest
```

```python
#Speaking of age lets do some basic analysis
#lets look at the number of total olympians by age
#Well.... I guess i cant use my age as an excuse anymore...
AgePlot= dataset[dataset['Age'].notna()]
AgePlot= sns.countplot(data=AgePlot,x='Age',hue='Sex',palette='Blues')
sns.set(rc={"figure.figsize":(20, 10)})
sns.set_style("whitegrid")
AgePlot.set(xlim=(2,30))

#Females dominate the 13-17 age category. Every other age category is male dominated 
```

```python
# lets take a look at what sports the females below 18 competed in
pd.set_option('display.float_format', lambda x: '%.2f' % x)
minimum=dataset.pivot_table(index=['Sport'], columns=['Sex'],
                            values='Age',aggfunc={'Age': ["min"]}, sort= False).plot.barh(figsize=(15,22))
minimum.legend(loc='upper right',fontsize=10)
```

```python
# lets take a look at what sports the females below 18 competed in
pd.set_option('display.float_format', lambda x: '%.2f' % x)
Maximum=dataset.pivot_table(index=['Sport'], columns=['Sex'],values='Age',aggfunc={'Age': ["max"]}, sort= True).plot.barh(figsize=(15,22))
Maximum.legend(loc='upper right',fontsize=10)
```

```python
# Filtering by the height to find out about the shortest Olympian, and we have two! Rosario Briones & Lyton Levison Mphande
Shortest=dataset.loc[dataset['Height'].idxmin()]
Shortest
```

```python
# Filtering by the height to find out about the tallest Olympian.
Tallest=dataset.loc[dataset['Height'].idxmax()]
Tallest
```

```python
# Filtering by the weight to find out about the lightest Olympian, which is Choi Myong-Hui
Lightest=dataset.loc[dataset['Weight'].idxmin()]
Lightest
```

```python
# Filtering by the weight to find out about the heaviest Olympian, which is Ricardo Blas, Jr. 
heaviest=dataset.loc[dataset['Weight'].idxmax()]
heaviest
```

```python
#Lets look at if there are any correlations in the data
# Looks like theres a strong correlation between height and weight 
correlation= dataset.dropna().corr()
correlation
```

```python
# lets visualize that im lazy 
sns.heatmap(correlation, annot=True)
```

```python
# lets vizualize that using a scatterplot 
# so basically,the taller you are, the more you weight
dataset.plot(kind='scatter', 
             x='Height',
             y='Weight')
plt.show()
```

```python
#Okay now lets look at sex,the number of females is a 1/3 of the number of males 
#But this is overall so lets look at it by year 
Sex=dataset['Sex'].value_counts().plot(kind='bar')
Sex.spines['bottom'].set_color('0')
Sex.spines['left'].set_color('0')
Sex.set_facecolor("white")
plt.xticks(rotation =360, fontsize=20)
plt.yticks( fontsize=20)
plt.xlabel('Sex',fontsize=20)
plt.grid(axis = 'y',color = 'grey', linestyle = '--', linewidth = 0.3)
```

```python
#Okay now lets see if that got better over the years
#Barely any females until 1912 sheeshhh..its getting better, not there yet.. maybe the male population is generally higher?
#A quick google search says that there are infact more men in the world
#but it looks like we have seasons?
ax= sns.countplot(x='Year',hue='Sex', data= dataset,palette='Blues')
sns.set(rc = {'figure.figsize':(24,10)})
ax.set_facecolor("white")
ax.spines['bottom'].set_color('0')
ax.spines['left'].set_color('0')
```

```python
#Lets get the exact count of males and females by year, no females in 1896 and only 16 in 1904? they mustve been amazing <3
#However, its lovely to see the gap reducing
dataset.groupby(['Year','Sex'])['Sex'].count()
```

```python
#Now lets investigate the seasonality
# so they only had summer seasons until 1920 and they've been alternating since 1994
# and obviously the summer season has more olympians participating 
dataset.pivot_table(index='Year', columns=['Season'],values='Sex',aggfunc='count' )
```

```python
# Lets explore the mdeals, Gold,Bronze and Silver 
dataset['Medal'].unique()
```

```python
# lets count the number of medals won by year and sex
SummerMedals=dataset[dataset['Season'].str.contains('Summer', na=False) & dataset['Medal'].str.contains('Gold',na=False)]
SummerMedals= SummerMedals.pivot_table(index='Year', columns=['Sex'],values='Medal',aggfunc='count' )
SummerMedals=SummerMedals.plot(kind='bar', figsize=(50,28))
SummerMedals.spines['bottom'].set_color('0')
SummerMedals.spines['left'].set_color('0')
SummerMedals.set_facecolor("white")
plt.xticks(rotation =360, fontsize=40)
plt.yticks( fontsize=40)
plt.xlabel('Age',fontsize=40)
SummerMedals.grid(axis = 'y',color = 'grey', linestyle = '--', linewidth = 0.3)
SummerMedals.legend(loc='upper right',fontsize=36)
```
